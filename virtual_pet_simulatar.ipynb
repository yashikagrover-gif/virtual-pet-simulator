```python
# Import required modules
import random
import json
```

```python
# FUNCTIONS 
# Function to display a random tip for the player

def show_tip():
    tips = [
        " Feed your pet before it gets too hungry.",
        " Playing increases happiness.",
        " Check your pet's status regularly.",
        " Balance food and play.",
        " Toys make your pet happier."
    ]

    print("\n" + random.choice(tips))

# Function to display the pet's current status

def show_status(pet, day):

    print("\n" + "=" * 45)
    print("🐾 PET STATUS")
    print("=" * 45)

    print("Pet Type   :", pet["emoji"], pet["type"])
    print("Name       :", pet["name"])
    print("Day        :", day)
    print("Hunger     :", pet["hunger"], "/100 ")
    print("Happiness  :", pet["happiness"], "/100 ")
    print("Energy     :", pet["energy"], "/100 ")

    print("XP         :", pet["xp"], "/100 ")
    print("Level      :", get_level(pet["xp"]))

    if pet["hunger"] >= 80:
        mood = "😫 Starving"

    elif pet["happiness"] >= 90:
        mood = "🥳 Excellent"

    elif pet["happiness"] >= 70:
        mood = "😄 Happy"

    elif pet["happiness"] >= 40:
        mood = "🙂 Fine"

    elif pet["happiness"] >= 20:
        mood = "😔 Sad"

    else:
        mood = "😭 Very Sad"

    print("Mood       :", mood)
    print("=" * 45)


# Function to feed the pet and update hunger, happiness and energy

def feed_pet(pet):
# Ask the user to choose a food item
    print("\n What should I eat today?")

    print("1. 🦴 Bone")
    print("2. 🍗 Chicken")
    print("3. 🐟 Fish")
    print("4. 🥛 Milk")
    print("5. 🍎 Apple")

    choice = input("Enter choice: ")

    if choice == "1":

        pet["hunger"] -= 15
        pet["happiness"] += 2
        pet["energy"] += 5

        print(f"{pet['emoji']} {pet['name']}: Yum! Bone was tasty! 😋")

    elif choice == "2":

        pet["hunger"] -= 25
        pet["happiness"] += 5
        pet["energy"] += 8

        print(f"{pet['emoji']} {pet['name']}: Chicken is my favourite! ❤️")

    elif choice == "3":

        pet["hunger"] -= 20
        pet["happiness"] += 8
        pet["energy"] += 6

        print(f"{pet['emoji']} {pet['name']}: Fish was delicious! 🐟")

    elif choice == "4":

        pet["hunger"] -= 10
        pet["happiness"] += 10
        pet["energy"] += 12

        print(f"{pet['emoji']} {pet['name']}: Milk made me happy! 🥛")

    elif choice == "5":

        pet["hunger"] -= 12
        pet["happiness"] += 6
        pet["energy"] += 4

        print(f"{pet['emoji']} {pet['name']}: Healthy choice! 🍎")

    else:
        print("❌ Invalid Choice!")

    pet["hunger"] = max(0, min(100, pet["hunger"]))
    pet["happiness"] = max(0, min(100, pet["happiness"]))
    pet["energy"] = max(0, min(100, pet["energy"]))

# Function to play with the pet and increase happiness

def play_pet(pet):
# Display different play activities
    
    print("\n🎮 Choose Activity")

    print("1. ⚽ Play Ball")
    print("2. 🚶 Go for Walk")
    print("3. 🎯 Play Fetch")
    print("4. 🫧 Bubble Game")

    choice = input("Enter choice: ")

    if choice == "1":

        pet["happiness"] += 15
        pet["hunger"] += 10
        pet["energy"] -= 15

    elif choice == "2":

        pet["happiness"] += 12
        pet["hunger"] += 8
        pet["energy"] -= 12

    elif choice == "3":

        pet["happiness"] += 18
        pet["hunger"] += 12
        pet["energy"] -= 18

    elif choice == "4":

        pet["happiness"] += 10
        pet["hunger"] += 5
        pet["energy"] -= 8

    else:

        print("Invalid Choice!")

    pet["hunger"] = min(100, pet["hunger"])
    pet["happiness"] = min(100, pet["happiness"])
    pet["energy"] = max(0, pet["energy"])

    messages = [
        "That was awesome! ",
        "Let's play again! ",
        "I'm so happy! ",
        "Best day ever! "
    ]

    print(f"{pet['emoji']} {pet['name']}: {random.choice(messages)}")

# Function to give a toy to the pet

def give_toy(pet):

    print("\n Choose a Toy")
    print("1. Teddy Bear 🧸")
    print("2. Ball ⚽")
    print("3. Rope 🪢")
    print("4. Frisbee 🥏")

    choice = input("Enter your choice: ")

    if choice == "1":
        pet["happiness"] += 12
        pet["energy"] -= 2
        print(f"{pet['emoji']} {pet['name']}: I love my Teddy! ❤️")

    elif choice == "2":
        pet["happiness"] += 10
        pet["energy"] -= 5
        print(f"{pet['emoji']} {pet['name']}: Let's play with the ball! ⚽")

    elif choice == "3":
        pet["happiness"] += 8
        pet["energy"] -= 4
        print(f"{pet['emoji']} {pet['name']}: Rope games are fun! ")

    elif choice == "4":
        pet["happiness"] += 15
        pet["energy"] -= 8
        print(f"{pet['emoji']} {pet['name']}: Catch the Frisbee! ")

    else:
        print(" Invalid Choice!")

    pet["happiness"] = min(100, pet["happiness"])
    pet["energy"] = max(0, pet["energy"])


# Function to give medicine and improve the pet's health

def give_medicine(pet):

    pet["hunger"] -= 5
    pet["happiness"] -= 2
    pet["energy"] += 10

    pet["hunger"] = max(0, pet["hunger"])
    pet["happiness"] = max(0, pet["happiness"])
    pet["energy"] = min(100, pet["energy"])

    print(f"{pet['emoji']} {pet['name']}: Yuck... medicine tastes bad! 🤢")
    print(f"{pet['emoji']} {pet['name']}: But I feel much better now! ❤️")

# Function to simulate the passing of time

def time_passes(pet):
# Hunger increases while happiness and energy decrease over time
    pet["hunger"] += 5
    pet["happiness"] -= 3
    pet["energy"] -= 2

    pet["hunger"] = min(100, pet["hunger"])
    pet["happiness"] = max(0, pet["happiness"])
    pet["energy"] = max(0, pet["energy"])

# Function to generate random events during the game

def random_event(pet):

    event = random.randint(1, 6)

    if event == 1:
        print(f"\n🎉 {pet['emoji']} {pet['name']} found a tasty snack!")
        pet["hunger"] -= 10

    elif event == 2:
        print(f"\n😴 {pet['emoji']} {pet['name']} took a peaceful nap.")
        pet["energy"] += 15
        pet["happiness"] += 5

    elif event == 3:
        print(f"\n🤒 {pet['emoji']} {pet['name']} caught a little cold.")
        pet["happiness"] -= 10
        pet["energy"] -= 10

    elif event == 4:
        print(f"\n🎈 {pet['emoji']} {pet['name']} made a new friend!")
        pet["happiness"] += 10

    elif event == 5:
        print(f"\n☀️ {pet['emoji']} {pet['name']} enjoyed the sunshine.")
        pet["happiness"] += 5
        pet["energy"] += 5

    else:
        print(f"\n🌧️ {pet['emoji']} {pet['name']} got wet in the rain.")
        pet["happiness"] -= 5

    pet["hunger"] = max(0, min(100, pet["hunger"]))
    pet["happiness"] = max(0, min(100, pet["happiness"]))
    pet["energy"] = max(0, min(100, pet["energy"]))

# Function to check whether the game should end

def game_over(pet):

    if pet["hunger"] >= 100:

        print("\n💀 GAME OVER!")
        print(f"{pet['emoji']} {pet['name']} became too hungry.")
        return True

    if pet["happiness"] <= 0:

        print("\n💀 GAME OVER!")
        print(f"{pet['emoji']} {pet['name']} became too sad.")
        return True

    if pet["energy"] <= 0:

        print("\n😴 Your pet is exhausted!")
        print("💤 Feeding or medicine can help restore energy.")

    return False

# Calculate Pet Level

def get_level(xp):

    if xp >= 100:
        return "🏆 Legendary Pet"

    elif xp >= 70:
        return "⭐⭐⭐ Super Pet"

    elif xp >= 40:
        return "⭐⭐ Active Pet"

    else:
        return "⭐ Beginner"

# Function to increase the pet's experience points

def gain_xp(pet, points):

    pet["xp"] += points

    if pet["xp"] > 100:
        pet["xp"] = 100

# Save Game

def save_game(pet, day, actions):

    import json

    game_data = {

        "pet": pet,

        "day": day,

        "actions": actions

    }

    with open("pet_save.json", "w") as file:

        json.dump(game_data, file, indent=4)

    print("\n Game Saved Successfully!")

# Function to load previously saved game data

def load_game():

    import json

    try:

        with open("pet_save.json", "r") as file:

            game_data = json.load(file)

        print("\n Saved Game Loaded Successfully!")

        return (
            game_data["pet"],
            game_data["day"],
            game_data["actions"]
        )

    except FileNotFoundError:

        print("\n No Saved Game Found!")

        return None, 1, 0
```

```python

# MAIN PROGRAM 

print("=" * 50)
print("🐾 WELCOME TO VIRTUAL PET SIMULATOR 🐾")
print("=" * 50)

print("\n1.  New Game")
print("2.  Load Saved Game")

start = input("\nChoose an option (1-2): ")

# Create a new game or load an existing one

if start == "2":

    pet, day, actions = load_game()

    if pet is None:
        start = "1"

# Allow the user to choose a pet

if start == "1":

    print("\nChoose Your Pet")

    print("1. 🐶 Dog")
    print("2. 🐱 Cat")
    print("3. 🐰 Rabbit")
    print("4. 🐦 Bird")

    pet_choice = input("Enter your choice (1-4): ")

    if pet_choice == "1":
        pet_type = "Dog"
        pet_emoji = "🐶"

    elif pet_choice == "2":
        pet_type = "Cat"
        pet_emoji = "🐱"

    elif pet_choice == "3":
        pet_type = "Rabbit"
        pet_emoji = "🐰"

    elif pet_choice == "4":
        pet_type = "Bird"
        pet_emoji = "🐦"

    else:
        print("Invalid choice! Default pet selected.")
        pet_type = "Dog"
        pet_emoji = "🐶"

    name = input("\nEnter your pet's name: ")

# Store pet details using a dictionary
    pet = {

        "name": name,

        "type": pet_type,

        "emoji": pet_emoji,

        "hunger": 50,

        "happiness": 50,

        "energy": 100,

        "xp": 0

    }

    day = 1

    actions = 0

# Main game loop

while True:

    print("\n" + "=" * 50)
    print("📅 Day", day)
    print("=" * 50)

    show_tip()
    
# Display the available actions
    
    print("\n========== MENU ==========")
    # Feed the pet
    print("1. 🍖 Feed Pet")
    # Play with the pet
    print("2. 🎮 Play with Pet")
    # Give the pet a toy
    print("3. 🧸 Give Toy")
    # Give medicine to the pet
    print("4. 💊 Give Medicine")
    # Display current pet status
    print("5. 📊 Check Status")
    # Save the current game progress
    print("6. 💾 Save Game")
    # Exit the game and display the final report
    print("7. 🚪 Quit")

    choice = input("\nEnter your choice: ")

    if choice == "1":
        feed_pet(pet)

        gain_xp(pet,10)

    elif choice == "2":

        play_pet(pet)

        gain_xp(pet,15)

    elif choice == "3":

        give_toy(pet)

        gain_xp(pet,5)

    elif choice == "4":

        give_medicine(pet)

        gain_xp(pet,5)
    elif choice == "5":

        show_status(pet, day)

        continue

    elif choice == "6":

        save_game(pet, day, actions)

        continue

    elif choice == "7":

        print("\n" + "=" * 50)
        print("🏆 FINAL REPORT")
        print("=" * 50)

        print("🐾 Pet Name      :", pet["name"])
        print("🐶 Pet Type      :", pet["emoji"], pet["type"])
        print("📅 Days Played   :", day)
        print("🍖 Hunger        :", pet["hunger"], "/100")
        print("😊 Happiness     :", pet["happiness"], "/100")
        print("⚡ Energy        :", pet["energy"], "/100")
        print("🌟 XP            :", pet["xp"], "/100")
        print("🏅 Level         :", get_level(pet["xp"]))

        if pet["happiness"] >= 80:
            print("\n⭐ Owner Rating : ⭐⭐⭐⭐⭐ Excellent")

        elif pet["happiness"] >= 60:
            print("\n⭐ Owner Rating : ⭐⭐⭐⭐ Very Good")

        elif pet["happiness"] >= 40:
            print("\n⭐ Owner Rating : ⭐⭐⭐ Good")

        else:
            print("\n⭐ Owner Rating : ⭐⭐ Needs Improvement")

        print("\n Thanks for taking care of your pet!")
        print(" Goodbye!")
        break

    else:

        print("\n Invalid Choice! Please select from 1 to 7.")
        continue

# Update pet attributes after each action
 
    time_passes(pet)
    
# Trigger a random event
   
    random_event(pet)
    
# Increase action count and update the day

    actions += 1

    if actions % 2 == 0:
        day += 1
# Check if the game should end

    if game_over(pet):
        break
```
